```python
import pandas as pd 
import os 
import requests
import json 
import matplotlib.pyplot as plt
```

```python
nypd_shootings=pd.read_csv('data/nypd.csv',encoding="ISO-8859-1")
nypd_shootings
```

```python
nypd_date=pd.concat([nypd_shootings.drop('OCCUR_DATE', axis = 1), 
          (nypd_shootings.OCCUR_DATE.str.split("/|T").str[:3].apply(pd.Series)
          .rename(columns={0:'month', 1:'day', 2:'year'}))], axis = 1)
nypd_date.head()
```

```python
nyc_shootings=nypd_date.rename(columns={'INCIDENT_KEY':'Incident Key',
                                       'BORO':'Borough','STATISTICAL_MURDER_FLAG':'Resulted in Murder',
                                       'year':'Year'})
nyc_shootings=nyc_shootings[['Incident Key','Borough','Resulted in Murder','Year']]
nyc_shootings.head()
```

```python
#brooklyn and queens shootings in 2008,2010,2016,2018
brooklyn_shootings_08=nyc_shootings[nyc_shootings['Borough']=='BROOKLYN'].groupby([nyc_shootings['Year']])
queens_shootings_08=nyc_shootings[nyc_shootings['Borough']=='QUEENS'].groupby([nyc_shootings['Year']])

```

```python
#total number of shootings in brooklyn 

bk_=brooklyn_shootings_08['Incident Key'].count()

#total number of shootings in queens 
qn_=queens_shootings_08['Incident Key'].count()
```

```python
bk_
```

```python
labels=['2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018']
plt.pie(bk_,autopct="%1.1f%%",labels=labels,shadow=True,startangle=150)
plt.title('% of Shootings in Brooklyn ')
plt.axis("equal")
plt.savefig('data/shootings.png')
plt.show()

```

```python
labels=['2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018']
plt.pie(qn_,autopct="%1.1f%%",labels=labels,shadow=True,startangle=150)
plt.title('% of Shootings in Queens ')
plt.axis("equal")
plt.savefig('data/shootingsqns.png')
plt.show()
```
