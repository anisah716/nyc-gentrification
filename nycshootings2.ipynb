```python
import pandas as pd 
import os 
import requests
import json 
import matplotlib.pyplot as plt
```

# Shooting Data 

```python
nypd_shootings=pd.read_csv('data/nypd.csv',encoding="ISO-8859-1")
nypd_shootings
```

```python
nypd_date=pd.concat([nypd_shootings.drop('OCCUR_DATE', axis = 1), 
          (nypd_shootings.OCCUR_DATE.str.split("/|T").str[:3].apply(pd.Series)
          .rename(columns={0:'month', 1:'day', 2:'year'}))], axis = 1)
nypd_date.head()
```

```python
nyc_shootings=nypd_date.rename(columns={'INCIDENT_KEY':'Incident Key',
                                       'BORO':'Borough','STATISTICAL_MURDER_FLAG':'Resulted in Murder',
                                       'year':'Year'})
nyc_shootings=nyc_shootings[['Incident Key','Borough','Resulted in Murder','Year']]
nyc_shootings.head()
```

# Total Shootings in Brooklyn 

```python
brooklyn_shootings=nyc_shootings[nyc_shootings['Borough']=='BROOKLYN'].groupby([nyc_shootings['Year']])
```

```python
#total shootings in brooklyn 
bk_total=brooklyn_shootings['Incident Key'].count()
bk_total
```

```python
brooklyn_chart = bk_total.plot(kind="bar", title="Total Number of Shootings in Brooklyn")
plt.xlabel("Years")
plt.ylabel("Number of Shootings")
plt.savefig('data/TotalShootingsinBrooklyn.png')
plt.show()
plt.tight_layout()
```

```python
queens_shootings=nyc_shootings[nyc_shootings['Borough']=='QUEENS'].groupby([nyc_shootings['Year']])
```

# Total Shootings in Queens
