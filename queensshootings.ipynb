```python
import pandas as pd 
import os 
import requests
import json 
import matplotlib.pyplot as plt
```

```python
nypd_shootings=pd.read_csv('data/nypd.csv',encoding="ISO-8859-1")
nypd_shootings
```

```python
nypd_date=pd.concat([nypd_shootings.drop('OCCUR_DATE', axis = 1), 
          (nypd_shootings.OCCUR_DATE.str.split("/|T").str[:3].apply(pd.Series)
          .rename(columns={0:'month', 1:'day', 2:'year'}))], axis = 1)
nypd_date.head()
```

```python
nyc_shootings=nypd_date.rename(columns={'INCIDENT_KEY':'Incident Key',
                                       'BORO':'Borough','STATISTICAL_MURDER_FLAG':'Resulted in Murder',
                                       'year':'Year'})
nyc_shootings=nyc_shootings[['Incident Key','Borough','Resulted in Murder','Year']]
nyc_shootings.head()
```

```python
queens_shootings=nyc_shootings.loc[nyc_shootings['Borough']=='QUEENS']
queens_shootings.head()
```

```python
qns_shootings=queens_shootings[['Incident Key','Resulted in Murder','Year']]
qns_shootings.head()
```

```python
qns_shoot=qns_shootings.groupby(['Year','Resulted in Murder'])

shoot_counts = qns_shoot.count()
shoot_counts
```

```python
chosen_year='2008'

qns_type=shoot_counts.loc[chosen_year]

labels=['Non-Fatal','Fatal']
plt.pie(qns_type,autopct="%1.1f%%",labels=labels,shadow=True,startangle=150)
plt.title('% Shootings in Queens in 2008')
plt.axis("equal")
plt.savefig('data/2008qnsshots.png')
plt.show
```
